## Exception Handling

- if something goes wrong
    - File that does not exits
    - Putting in wrong `input()` e.g `int(input("abcd"))`
    - Network errors - accessing links or network

- Handles exceptions and says to python:
    "handle the code if there is an error, like this"

- So instead of Crashing / Bombing - Our program executes none the less

```py

try: 
    # code that might give errors

except:

    # what to do for an error

### File Errors:
- file not found type of errors


````python
filename = 'idrees.txt'

try:
    with open(filename, "r") as f:
        print(f.read())
        
except FileNotFoundError:
    print(f'Error: {filename} does not EXIST!! YOU DUMMY!! check the files in the folder again!!')
except Exception as e:
    print("Something wrong. Check stuff again. ERRORRRRRR:::: ", e)
````

### Line Counter Program
- Create a line counter program
- Take a file name
- Read the lines
- Print the line count
- Handle Errors if file does not exist, or filename is wrong

> (10 Minutes - Best of Luck!!)

````python
def line_counter():
    
    line_count = 0
    try:
        filename = input("Enter File Name: ")  

        with open(filename, "r") as file:
            lines = file.readlines()
            line_count = 0
            for i in lines:
                line_count += 1 
        print("Name of the Document:", filename)
        print("Number of Lines in the Document:", line_count)

    except FileNotFoundError:
        print("File Not Found. Check the Name again.")
    except Exception as e:
        print("Something wrong. Check input: ", e)    

    return line_count
````
