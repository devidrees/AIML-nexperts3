### Selection by:
- Location  --- `.iloc[ ]`
- selecting by integer location `i` - integer 
- 
- `df.iloc[row index]` # access rows
- `df.iloc[row index, column index]` # access values at = row , column
- `df.iloc[row_start:row_end , col_start:col_end ]` # access === range of rows, range of columns



````python
df.iloc[:] # works --- gives evrything back
df.iloc[4] # returns a series object
# header turned to index
# getting values for different rows

# slicing will work too
df.iloc[:3]             # first 3 rows

# can we access a limited number of columns too?
# INDEXING

df.iloc[:,:]
df.iloc[:3,:2] # first 3 rows, first 2 columns

df.iloc[0:3, :3] # 0 1 2
df.iloc[0:3, :4] # 0 1 2 3
df.iloc[0:3, 1:2] # 1 

df.iloc[3 , 0 ] # value =  row 3 , column 1  


# we can get any subset of a dataframe like this
````

**.iloc**
- Label --- `.loc [  ]`
- we go to a location using a `label`
- `df.loc [row_label]` - return a value from that row
- `df.loc[row_label]` # access rows
- `df.loc[row_label, column_label]` # access values at = row , column
- `df.loc[row_start_label:row_end_label , col_start_label:col_end_label ]` # access === range of rows, range of columns

````python
df.loc[4] # target rows
````

````python
data = {
    'Student': ['Idrees','Saufi','Faiz','Dila','Nadirah'],
    'Math': [99, 85, 95, 85, 98],
    'English': [45, 35, 25, 65, 59],
    'Science': [55, 65, 75, 85, 87],
}

df = pd.DataFrame(data, index = ['stu1','stu2', 'stu3', 'stu4', 'stu5'])
````

````python
df
````

````python
df.iloc[4] # iloc still works----- because it target integer locations

#df.loc[5] # won't work
#df.iloc['stu1'] # won't' work

df.loc[:'stu4'] # get all the rows upto stu4

df.loc['stu1': 'stu4', 'Math':'Science'] # setup a range too
````

- Selection based on DTypes - `df.select_dtypes(include=['number', 'object', 'datetime','bool','str'])`

- Selection using List Comprehension - `cols = [ condition ]` ---> 'df[cols]'

```py 
    [(operation on x) for x in <iterable> (condition)]

````python
# Revision of ----- List Comprehension
[x for x in range(10)]

[x for x in range(10) if x % 2]

[(3*x + 2) for x in range(10) if x % 2] # [(operation on x) for x in <iterable> (condition)]

````
