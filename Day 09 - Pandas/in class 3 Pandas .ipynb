```python
import pandas as pd
data = {
    'Student': ['Idrees','Saufi','Faiz','Dila','Nadirah'],
    'Math': [99, 85, 95, 85, 98],
    'English': [45, 35, 25, 65, 59],
    'Science': [55, 65, 75, 85, 87],
}

df = pd.DataFrame(data)
```

### Aggregation, Grouping, Filtering 

- Aggregation Methods in `Pandas`:
    - can apply aggregations on series objects
    - `['mean','max','min', 'count', 'sum', 'std', 'var','prod']`
    - std - standard deviation
    - var - variance
    - prod - product
    - 
    - `series.agg([cols])` - multiple aggregations on the same column

```python
# aggregation - combine stuff to get some metrics
df.English.mean() # can apply 


df['Math'].mean()

df.Science.count() # how many rows there are in a column
df.Science.value_counts()
```

```python
# multiple aggregations on the SAME column

df.Science.agg(['mean','max','min', 'count', 'sum', 'std', 'var','prod'])
```

```python
# aggregations for MULTIPLE columns

df[['Math','English','Science']].mean()
df[['Math','English','Science']].sum()
df[['Math','English','Science']].min()
df[['Math','English','Science']].max()
df[['Math','English','Science']].count()
df[['Math','English','Science']].value_counts()
```

## Grouping in pandas

We keep some catagories in data

```python
data = {
    'Student': ['Idrees','Saufi','Faiz','Dila','Nadirah'],
    'Liscence': ['Yes', 'No', 'Yes', 'No', 'No'],
    'Age': [30, 30, 30, 18, 18],
    'F_subject': ['Science', 'Science', 'English','Physics', 'Physics']
}

df_a = pd.DataFrame(data)
```

```python
# aggregates for a category
df_a.groupby('F_subject')['Age'].sum() 

df_a.groupby('Student')['F_subject'].count()
```
