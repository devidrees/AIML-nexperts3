#### BarCharts in matplotlib
- Bar chart will be used for a catagorical variable
- all teh arguments that go into the `plt.bar()`
- univariate, bivariate - multivariate
- 

```python
products = ["Apple", "Mango", "Orange", "Banana"]
sales = [10 , 25, 45, 30]

# to make things easy
x = ["Apple", "Mango", "Orange", "Banana"]
y = [10 , 25, 45, 30]

import matplotlib.pyplot as plt

# plt.bar(x, y )
plt.bar(products, sales)
```

```python
# we get a lot of arguments with these methods
# some commonly used ones
# use whichever you need in any case


plt.bar(
    x= products,        # horizontal varaible
    height=y,           # height or vertical magnitudes
    width= 0.5,         # width of the bar
    bottom = None,      # stacked bar charts
    align= 'center',    # center your viz
    color = "#078776",     # use to change colors, # can use hex
    edgecolor = "#104B4B", # border color
    linewidth = 2,           # border thickness
    tick_label = ['a','b','c','d'],  # names for each bar
    hatch = '////\\' # x /\\ + o O
)


patterns = ['/', 'o', '*', '\\']
plt.bar(products, sales, hatch =patterns)
```

```python
x = ['Group A', 'Group B', 'Group C']
y = [20, 35, 30]
hatches = ['---', '+++', 'xxx']
colors = ['#f0f0f0', "#7b1010", "#491C93"] # Light to dark gray

plt.bar(x, y, color=colors, edgecolor='black', hatch=hatches)
plt.show()
```

```python
# Stacker Bar charts
x = ["Apple", "Mango", "Orange", "Banana"]
y = [10 , 25, 45, 30]
y1 = [10 , 25, 45, 30]

plt.bar(x,y, label = 'Online', color = "#15E654")

# we can do bi-variate analysis
plt.bar(x, y1, label = 'y1', bottom = y , color = "#9D2191")
```

```python
# multivariate analysis
import numpy as np

x = ["Apple", "Mango", "Orange", "Banana"]
y = np.array([10 , 25, 45, 30])
y1 = np.array([10 , 25, 45, 30])
y2 = np.array([10 , 25, 45, 30])

# we can do multi-variate analysis
plt.bar(x, y, label = 'y', bottom = y1 + y2, color = "#9D2191")
```

#### Scatter plot
- 

```python
x = [1,2,3,4,5,6,7]
y = [21,100,35,25,22,15,23]
bubble_sizes = [100, 200, 300, 400, 500, 600, 700 ]
y2 = [22,55,33,43,26,32,28]

plt.scatter(x, y, 
            s = bubble_sizes, # size of each point in scatterplot
            c = y,   # color
            cmap= 'inferno', # color map 
            alpha=0.6, 
            marker= '^', # x, +, *, v , ^
            edgecolors='black', 
            linewidths=2)

plt.scatter(x, y2, s = bubble_sizes, c = y2, cmap='viridis', marker= 'o', alpha=0.5) # inferno plasma magma viridis
```
