### $Seaborn$

- a powerful Python library built on top of Matplotlib, designed for creating attractive and informative statistical plots


```python
# pip install seaborn
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_iris = sns.load_dataset('iris')
df_iris.head(2)

df_penguins = sns.load_dataset('penguins')
df_penguins.head(2)

df_titanic = sns.load_dataset('titanic')
df_titanic.head(2)
```

```python
# check available dataset names
# sns.get_dataset_names()
```

```python
# check color palette
sns.color_palette()
```

```python
# set a theme for visualizations
sns.set_theme()
```

Styles
- darkgrid – Light background with gray gridlines (default).
- whitegrid – White background with subtle gridlines.
- dark – Dark background without grids.
- white – White background without grids.
- ticks – White background with light ticks.

Palettes: [link](https://www.practicalpythonfordatascience.com/ap_seaborn_palette)
- Qualitative Palettes: Good for representing categorical data. Examples include "deep", "muted", "bright", "pastel", "dark", and "colorblind".
- Sequential Palettes: Good for representing numeric data. Examples include "rocket", "mako", "flare", and "crest".
- Diverging Palettes: Good for representing numeric data with a categorical boundary. Examples include "vlag" and "icefire".



```python
# Apply a theme
sns.set_theme(style="darkgrid", palette="deep")
```

```python
# bar plot
sns.barplot(
    data= df_titanic,
    x = 'pclass',
    y = 'survived',
    hue= 'sex',
    palette= 'dark',     # is affecting the bars
    capsize= 0.1,
    orient= 'h',          # v and h to orient graphs
    edgecolor = '#000000',
    linewidth = 2
            )
```

```python
plt.figure(figsize= (5,3))

sns.barplot(data= df_titanic,x = 'pclass',y = 'survived',hue= 'sex')

plt.title("Survivor's Class")

plt.xlabel('Class')
plt.ylabel('Survivors')

plt.show()
```

```python
# countplot
# helps us to estimate the count of observations
plt.figure(figsize= (5,3))
sns.countplot(
    data = df_titanic,
    x = 'pclass',
    hue = 'sex',
    alpha = 0.8,
    palette= 'bright',
    edgecolor = "#F423B6",
    linewidth = 2
)
```

```python
# scatter plot
plt.figure(figsize= (5,3))
sns.scatterplot(
    data = df_iris,
    x = 'petal_length',
    y = 'sepal_length',
    hue= 'species',
    palette= 'viridis',
    alpha = 0.7,
    legend = 'auto'
)

plt.title("Sepal and petal length plot")
plt.xlabel('petal length')
plt.ylabel('sepal length')

plt.show()
```

```python
# regplot: regression plot
plt.figure(figsize= (5,3))
sns.regplot(
    data = df_iris,
    x = 'petal_length',
    y = 'sepal_length',
    scatter_kws= {'alpha':0.3},
    line_kws= {'color': 'r'}
    )
```

```python
# line plot
# create plots of lines
plt.figure(figsize= (5,3))
sns.lineplot(
    data = df_titanic,
    x = 'age',
    y = 'fare'
)
```

```python
import numpy as np

x = [x for x in range(10)]
y  = [y for y in np.random.randint(20, 300, 10)]
y2  = [y for y in np.random.randint(1, 4, 10)]

plt.figure(figsize= (5,3))
sns.lineplot(
    x = x,
    y = y,
    hue = y2,   # different colors for different lines
    style = y2,  # different dashing for different lines
    markers= True,
    palette= 'dark'

)

```

```python
# pie chart

plt.figure(figsize= (5,3))

# Define data and labels
data = [15, 25, 25, 30, 5]
labels = ["Apple", "Mango", "Orange", "Banana", "Pineapple"]

# Use Seaborn color palette
colors = sns.color_palette('mako')[0:5] # 

# Create pie chart
plt.pie(data, 
        labels=labels, 
        colors=colors, 
        autopct='%.0f%%')

plt.show()
```

```python
# pi chart of the Passesnger class

plt.figure(figsize= (5,3))
colors = sns.color_palette('mako')[0:3] # three colors# draw circle

plt.pie(
    df_titanic.pclass.value_counts(),
    labels = ['Class 3', 'Class 1', 'Class 2'],
    autopct= '%1.1f%%',
    colors = sns.color_palette('mako')
)

plt.show()
```

```python
plt.figure(figsize= (5,5))
plt.gcf().gca().add_artist(plt.Circle((0.5, 0.5), 0.3, fc='white'))
# creates a white circle at centre
```

```python
plt.pie(
    df_titanic.pclass.value_counts(),
    labels = ['Class 3', 'Class 1', 'Class 2'],
    autopct= '%1.1f%%',
    colors=["#16DC00","#13BA00","#0E8800"],
    pctdistance= 0.7,
)

plt.gcf().gca().add_artist(plt.Circle((0, 0), 0.5, fc='white'))
plt.title('Donut Chart')

```

```python
# draw circle
centre_circle = plt.Circle( # create a circle object
    (0, 0),             # centre of the circle
    0.70,               # radius
    fc='white')         # facecolor
fig = plt.gcf()         # get current figure 

# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)    # Get Current Axes 
```

```python
# histogram
sns.histplot(
    data= df_titanic,
    x = 'age',
    bins=30,     # divide data into this many bars
    kde=True,     # kernel density estimate - shows a smoother line for distribution
    hue = 'survived',
    palette= 'bright',
    color= 'b',
    element= 'step'    # bars, poly, step
)

plt.title('Histogram')
```

```python
# Boxplot
# in seaborn will work with one column, or multiple as well

sns.boxplot(
    data= df_titanic,
    x = 'pclass',
    y = 'age',
    hue= 'survived',
    width= 0.8,
    fliersize= 5, # control the size fo the outlier dots
)
```

```python
# pairplot
plt.figure(figsize= (4,4))
sns.pairplot(
    data = df_titanic[['survived','age','fare']],
    hue = 'survived',
    palette= 'husl',
    diag_kind = 'kde'
)
```

```python

# plt.figure(figsize= (4,4),dpi=720)
# create pairplot for iris?
sns.pairplot(df_iris, hue="species", palette="husl",  # figure level function ----- creates its own fig or canvas
             height= 2, aspect=1)
plt.tight_layout()
plt.show()
```

```python
# correlation maps --------- OPTIONAL
num_titanic = df_titanic.select_dtypes(include=['number'])

corr = num_titanic.corr()    
# adjacency matrix ----- each variable on the left, and the top as well

sns.heatmap(
    corr,
    annot= True, # annotations
    cmap= 'vlag', # inferno vlag aquamarine viridis
    fmt= ".1f", # round decimal places
)
```

```python
# Dashboard example

# set theme
my_palette = sns.color_palette('pastel', as_cmap= True)
sns.set_theme(style='darkgrid', palette = my_palette)


# create the canvas or figure or grid
fig, ax = plt.subplots(2,2, figsize = (12,8))





# viz 1-------ScatterPlot
sns.scatterplot(
    data = df_iris,
    x = 'petal_length',
    y = 'sepal_length',
    hue= 'species',
    alpha = 0.7,
    legend = 'auto',
    ax=ax[0,0]
)
ax[0,0].set_title("Scatter Plot",fontsize = 20,)








# viz 2-----------PIE CHART
ax[0,1].pie(
    df_titanic.pclass.value_counts(),
    labels = ['Class 3', 'Class 1', 'Class 2'],
    autopct= '%1.1f%%',
    pctdistance = 0.80,
    
)
centred_circle = plt.Circle((0, 0), 0.5, fc='white')
ax[0,1].add_artist(centred_circle)
ax[0,1].set_title("Donut Chart", fontsize = 20,)








# viz 3--------HEATMAP
num_titanic = df_titanic.select_dtypes(include=['number'])
corr = num_titanic.corr() 
sns.heatmap(
    corr,
    annot= True, # annotations
    fmt= ".1f", # round decimal places,
    cmap= my_palette,
    ax = ax[1,0]
)
ax[1,0].set_title("HEAT MAP", fontsize = 20,)








# viz 4-----------HISTOGRAM
sns.histplot(
    data= df_titanic,
    x = 'age',
    bins=30,     # divide data into this many bars
    kde=True,     # kernel density estimate - shows a smoother line for distribution
    hue = 'survived',
    element= 'step',    # bars, poly, step
    palette= my_palette,
    ax = ax[1,1]
)
ax[1,1].set_title("HISTOGRAM", fontsize = 20,)









# final touchup
plt.tight_layout()
plt.suptitle('Dashboard for Analysis',
             fontsize = 42,
             fontweight = 'bold',
            y = 1.1)
```
