```python
import plotly.express as px

# px object --- is wrapper that takes dataframe and creates a chart

# Load datasets from plotly --- inbuilt

df = px.data.iris()
```

```python
fig = px.scatter(
    df,
    x = 'sepal_length',
    y= 'petal_length',
    size = 'petal_width',
    hover_data= ['species','petal_width'],
    hover_name= 'species',                  # HOVER BOX - name you can specify here 
    color= 'species',
    facet_col= 'species'                    # create subplot, without creating subplots!!
    )

fig.show()
```

```python
import plotly.express as px
import pandas as pd

# Sample data
data = {
    "Fruits": ["Apples", "Bananas", "Cherries", "Dates"],
    "Quantity": [10, 15, 7, 12]
}

df = pd.DataFrame(data)

# Create bar chart
fig = px.bar(
    df,
    x="Fruits",
    y="Quantity",
    title="Fruit Quantity Bar Chart",
    color="Fruits",  # Adds color by category
    text="Quantity"  # Show values on bars
)

fig.show()
```
